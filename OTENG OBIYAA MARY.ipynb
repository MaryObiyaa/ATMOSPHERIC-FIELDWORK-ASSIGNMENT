```python
import numpy as np

# Sample observational data (in degrees Celsius)
observations = np.array([15.2, 16.1, 14.5, 15.8, 25.0])

# Sample model predictions (in degrees Celsius)
model_output = np.array([14.8, 15.5, 14.0, 16.0, 25.3])

# Observation error variance (assumed constant for simplicity)
obs_error_var = 1.0

# Background error variance (assumed constant for simplicity)
bg_error_var = 4.0

# Weighting factor based on error variances
weights = obs_error_var / (obs_error_var + bg_error_var)

# Perform Optimal Interpolation (OI)
analysis = weights * observations + (1 - weights) * model_output

# Print analysis results
print("Analysis:")
print(analysis)

# Calculate root mean square error (RMSE)
def rmse(a, b):
  return np.sqrt(np.mean((a - b) ** 2))

rmse_obs = rmse(analysis, observations)
rmse_model = rmse(analysis, model_output)

# Print RMSE values
print("\nRMSE:")
print("  Observations:", rmse_obs)
print("  Model Output:", rmse_model)

# Calculate bias
bias_obs = np.mean(analysis - observations)
bias_model = np.mean(analysis - model_output)

# Print bias values
print("\nBias:")
print("  Observations:", bias_obs)
print("  Model Output:", bias_model)

#Absolutely, here's the code with an enhanced visualization using Matplotlib:
import numpy as np
import matplotlib.pyplot as plt

# Sample observational data (in degrees Celsius)
observations = np.array([15.2, 16.1, 14.5, 15.8, 25.0])

# Sample model predictions (in degrees Celsius)
model_output = np.array([14.8, 15.5, 14.0, 16.0, 25.3])

# Observation error variance (assumed constant for simplicity)
obs_error_var = 1.0

# Background error variance (assumed constant for simplicity)
bg_error_var = 4.0

# Weighting factor based on error variances
weights = obs_error_var / (obs_error_var + bg_error_var)

# Perform Optimal Interpolation (OI)
analysis = weights * observations + (1 - weights) * model_output

# Data labels
data_labels = ["Observation", "Model Output", "Analysis"]


plt.figure(figsize=(15, 10))
plt.plot(observations,  label='Observations otput', color='Yellow')
plt.plot(model_output, label='Model_Output', color='green')
plt.plot(analysis, label='Analysis', color='blue')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('OPTIMAL INTERPOLATION')
plt.legend()
plt.grid(True)
plt.show()
```
